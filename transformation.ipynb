```python
import pandas as pd

# Load the EHR dataset
ehr_df = pd.read_csv('EHR_cleaned.csv')

# Load the PAMAP2 dataset
wearable_df = pd.read_csv('wearable_cleaned.csv', delimiter=',')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of time spent in the hospital
plt.figure(figsize=(10, 6))
sns.histplot(ehr_df['time_in_hospital'], bins=15, kde=True)
plt.title('Distribution of Time Spent in Hospital')
plt.xlabel('Time in Hospital (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of readmission rates by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='readmitted', hue='gender', data=ehr_df, palette='Set2')
plt.title('Readmission Rates by Gender')
plt.xlabel('Readmission Status')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot of average time in hospital by age group
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='time_in_hospital', data=ehr_df, estimator='mean', ci=None, palette='coolwarm')
plt.title('Average Time in Hospital by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Time in Hospital (Days)')
plt.show()
```

```python
# Distribution of heart rates in the wearable dataset
plt.figure(figsize=(10, 6))
sns.histplot(wearable_df['heart_rate'], bins=30, kde=True)
plt.title('Distribution of Heart Rates (Wearable Data)')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Distribution of hand accelerometer (y-axis) data
plt.figure(figsize=(10, 6))
sns.histplot(wearable_df['hand_acc_16_y'], bins=30, kde=True)
plt.title('Distribution of Hand Accelerometer Data (Y-axis)')
plt.xlabel('Hand Acc 16 Y')
plt.ylabel('Frequency')
plt.show()
```
